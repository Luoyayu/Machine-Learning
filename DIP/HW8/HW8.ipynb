### Shannon-Fano编码

1. 0 4 5 6 7 -> (0 4) | (5 6 7) 或 (4 7) | (0 5 6)

    将0,4,5,6,7 拆分成0,4和5,6,7 或4,7和0,5,6 使得拆分后的两组频数差最小

    讨论两种情况

2. 1) 若是(0  4) | (5 6 7) -> (0) (4) | (5) | (6 7)

    若是情况一将0,4,5,6,7拆分成(0,4)和(5,6,7) 其频数和差最小，此时编码(0,4)为0, (5,6,7)为1

    继续对(0,4)和(5,6,7)拆分，进一步得到下图编码方案

| 符号 |  0   |  4   |  5   |  6   |  7   |
| :--:| :--: | :--: | :--: | :--: | :--: |
| 编码 |  00  |  01  |  10  | 110  | 111  |

平均比特数 $\displaystyle \frac{2\mathrm{Bit}*(6+28+18)+3\mathrm{Bit}*（10+2）}{64\mathrm{Symblo}}=2.1875\mathrm{\ Bits \ per \ Symbol}$

2) 若是(4  7) | (0 5 6) -> (4) (7) | (0) (6) |(5)

按照情况一的方法拆分(0,4,5,6,7), 得到下图的编码的方案

| 符号 |  0   |  4   |  5   |  6   |  7   |
| :--:| :--: | :--: | :--: | :--: | :--: |
| 编码 |  110  |  00  |  10  | 111 | 01 |

平均比特数 $$\displaystyle \frac{2\mathrm{Bit}*(28+18+2)+3\mathrm{Bit}*（6+10）}{64\mathrm{Symblo}}=2.25\mathrm{\ Bits \ per \ Symbol}$$ 

可见情况一的编码更优，计算其编码效率 $\eta$为信息熵/平均比特数

编码效率：$-(0.09375*log2 0.09375 + 0.4375*log2 0.4375 + 0.28125*log2 0.28125 \\ +0.15625*log2 0.15625 +  0.03125*log2 0.03125)=1.93134931568$  

$\eta = 1.93134931568/2.1875=0.88290254431$

#### 编码方案对比

霍夫曼编码使用变长编码表对源符号进行编码，这便使编码之后的字符串的平均长度、期望值降低，    
霍夫曼使用自底向上的方法构建二叉树，避免了次优算法香农-范诺编码（Shannon–Fano coding）的最大弊端──自顶向下构建树。  
霍夫曼编码是最优编码方案，而通过香农-范诺编码得出的未必是最佳编码。
