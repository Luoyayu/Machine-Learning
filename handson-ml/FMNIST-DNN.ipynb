```python
# ================================================================================================
# Author: Tongyuhu
# MLP apply in Fashion-MNIST classification base on TensorFlow High-Level API
# two hidden layers; 1st: 300 neurons; 2ed: 100 neurons
# Softmax output layer with 10 neurous

import numpy as np 
import tensorflow as tf
from sklearn.datasets import fetch_mldata
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler

(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()# 使用keras后端调取数据
X_train = X_train.astype(np.float32).reshape(-1, 28*28) / 255.0
X_test = X_test.astype(np.float32).reshape(-1, 28*28) / 255.0
y_train = y_train.astype(np.int32)
y_test = y_test.astype(np.int32)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=1/6)

print('Train Set', X_train.shape, y_train.shape)
print('Valid Set', X_val.shape, y_val.shape)
```

```python
%%time
# ================================================================================================
feature_cols = [tf.feature_column.numeric_column("X", shape=[28*28])]

dnn_clf = tf.estimator.DNNClassifier(hidden_units=[300,300], n_classes=10, 
                                    feature_columns=feature_cols)
input_fn = tf.estimator.inputs.numpy_input_fn(
        x={'X':X_train}, y=y_train, num_epochs=80, batch_size=50, shuffle=True)

dnn_clf.train(input_fn=input_fn)
```

```python
%%time

test_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={'X': X_test}, y = y_test, shuffle=True
)
eval_result = dnn_clf.evaluate(input_fn=test_input_fn)
print(eval_result)
```

```python
%%time
val_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={'X': X_val}, y = y_val, shuffle=True
)
val_eval_result = dnn_clf.evaluate(input_fn=val_input_fn)
print(val_eval_result)
```
